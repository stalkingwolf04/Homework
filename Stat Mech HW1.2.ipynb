```python
import numpy as np
import pandas as pd
import math
```

```python
N_total = 200  #total number of paramagnets
N_A = 100  # number of paramagnets in A
N_B = 100  # number of paramagnets in B
q_total = 80  # number of energy units in the system

N_up_total = N_total - q_total  #number of up paramagnets in total system
N_A_down = np.arange(0,81)  # will increment this from 0->80
```

```python
mult_system = math.factorial(N_total)/(math.factorial(N_up_total)*math.factorial(q_total))
# checked with calculator - OK
mult_A = np.zeros(81)
mult_B = np.zeros(81)
m_combined = np.zeros(81)
prob = np.zeros(81)
```

```python
N_A_up = N_A - N_A_down
N_B_down = q_total - N_A_down
N_B_up = N_B - N_B_down

for i in range(0,81):
    mult_A[i] = np.math.factorial(N_A)/(np.math.factorial(100-i)*np.math.factorial(i))
    mult_B[i] = np.math.factorial(N_B)/(np.math.factorial(100-(80-i))*np.math.factorial(80-i))
    m_combined = mult_A*mult_B
    prob = m_combined/mult_system
```
